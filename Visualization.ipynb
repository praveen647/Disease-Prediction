
##
## Piechart for disease distribution

```python
import seaborn as sns
import matplotlib.pyplot as plt
df_train = pd.read_csv("Training.csv")
df_test = pd.read_csv("Testing.csv")
df_full = pd.concat([df_train,df_test])
df_full = df_full.iloc[:,:-1]
df_full.rename(columns={"prognosis":"disease"},inplace=True)
df_full['disease'].value_counts().plot(kind='pie', autopct='%1.1f%%', figsize=(17, 17))
plt.title('Disease Distribution')
plt.ylabel('')
plt.show()
```

## 

## Bar-plot for Total number of Symptoms per Disease

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
symptoms['total_symptoms'] = symptoms.sum(axis=1)
df_dis = pd.concat([symptoms['total_symptoms'], pd.Series(disease,name="disease")], axis=1)

# Group by disease and sum the total symptoms for each disease
symptom_totals = df_dis.groupby('disease')['total_symptoms'].sum().reset_index()

# Sort values to make plot more readable
symptom_totals = symptom_totals.sort_values(by='total_symptoms', ascending=False)

# Plot the total number of symptoms per unique disease
plt.figure(figsize=(10, 10))
sns.barplot(x='total_symptoms', y='disease', data=symptom_totals, palette="Blues_d")
plt.title('Total Number of Symptoms per Disease')
plt.xlabel('Total Number of Symptoms')
plt.ylabel('Disease')
plt.show()
```

##
## Bar-plot for Symptom Distribution

```python
feature_importance = symptoms.drop(['total_symptoms'], axis=1).sum().sort_values(ascending=False)
feature_importance.plot(kind='bar', figsize=(30,10))
plt.title('Importance of Each Symptom')
plt.ylabel('Symptom Count')
plt.xticks(rotation=90)
plt.show()
```

##
## Bar-plot for Most Frequent Symptom per Disease

```python

# Assuming 'symptoms' is your dataframe with binary symptom columns
# 'disease' is your dataframe with one column corresponding to the disease

# Concatenate symptoms with disease
# df = pd.concat([symptoms, disease], axis=1)

# Group by disease and sum the occurrences of each symptom
symptom_sums = df_full.groupby('disease').sum()

# Find the most frequently occurring symptom for each disease
most_frequent_symptom = symptom_sums.idxmax(axis=1).reset_index()
most_frequent_symptom.columns = ['disease', 'most_frequent_symptom']

# Merge the most frequent symptom back with its occurrence count
symptom_counts = symptom_sums.max(axis=1).reset_index()
symptom_counts.columns = ['disease', 'max_occurrence']

# Combine both results into one dataframe
most_frequent_symptom = pd.merge(most_frequent_symptom, symptom_counts, on='disease')

# Plot the most frequent symptom for each disease
# Plot the most frequent symptom for each disease
# Plot the most frequent symptom for each disease
# Plot the most frequent symptom for each disease
# Plot the most frequent symptom for each disease
plt.figure(figsize=(12, 6))
sns.barplot(x='disease', y='max_occurrence', hue='most_frequent_symptom', data=most_frequent_symptom, dodge=False)
plt.title('Most Frequent Symptom per Disease')
plt.ylabel('Number of Occurrences of Most Frequent Symptom')

# Reduce the size of the x-axis labels
plt.xticks(rotation=45, ha='right', fontsize=7)  # Adjust fontsize to make x-axis labels smaller

plt.tight_layout()

# Move the legend to the bottom and adjust its size
plt.legend(
    title='Most Frequent Symptom', 
    loc='upper center', 
    bbox_to_anchor=(0.5, -0.25),  # Move the legend a bit lower
    ncol=4,  # Adjust the number of columns to make it wider
    fontsize='x-small',  # Make the hue text smaller
    title_fontsize='small',  # Make the title slightly larger than the legend items
    handletextpad=0.3  # Reduce space between legend markers and text
)

# Show the plot
plt.show()



```

## Histogram for Feature Importance

```python
from sklearn.ensemble import RandomForestClassifier

# Assuming 'X' as the features and 'y' as the target column
X = df_full.drop(columns=['disease'])  # Features
y = df_full['disease']  # Target

model = RandomForestClassifier()
model.fit(X, y)

# Plot Feature Importances
importances = model.feature_importances_
indices = np.argsort(importances)[-15:]  # Top 15 features

plt.figure(figsize=(10, 6))
plt.title('Feature Importance (Top 15)')
plt.barh(range(len(indices)), importances[indices], color='b', align='center')
plt.yticks(range(len(indices)), [X.columns[i] for i in indices])
plt.xlabel('Relative Importance')
plt.show()
```
